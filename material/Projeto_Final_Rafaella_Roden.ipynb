## Remuneração x Ano

*Pergunta: Existiu uma queda na remuneração dos profissionais de dados considerando os anos analisados?*

```python
#qual a contagem
salario_ano_contagem = df.groupby('Ano')['Faixa salarial'].value_counts()
salario_ano_contagem
```

```python
#qual a proporção de faixa salarial por ano.
salario_ano_proporcao = df.groupby('Ano')["Faixa salarial"].value_counts(normalize=True).reset_index(name='Proporcao')
salario_ano_proporcao
```

```python
#alterando a faixa salarial para string e assim conseguir usar no gráfico
salario_ano_proporcao['Faixa salarial'] = salario_ano_proporcao['Faixa salarial'].astype(str)
```

```python
# importando os pacotes que iremos utilizar
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#pivatando = montar a tabela primeiro para depois criar a visualização.
mapa_calor = pd.pivot_table(salario_ano_proporcao, values='Proporcao', index='Faixa salarial', columns='Ano')
```

```python
sns.heatmap(mapa_calor, annot=True, fmt=".2f", cmap="Greens")
plt.title("Salários por Faixa Salarial e Ano")
plt.show()
```

Aparentemente não temos alteração significava em remuneração para quem já está atuando.

## Remuneração x Atuacao

```python
#verificando a faixa salarial por atuação
salario_cargo = df.groupby(['Atuacao', 'Ano','Faixa salarial']).size() / df.groupby('Ano').size()
salario_cargo=salario_cargo.reset_index(name='Proporcao')
salario_cargo
```

```python
#grafico que mostra a proporção de pessoas em cada área de atuação por ano
sns.barplot(x='Ano', y='Proporcao', hue='Atuacao', data=salario_cargo, dodge=True)
plt.show()
```

Mais profissionais atuando como analista de dados e menos cientistas de dados.


```python
#verificando a faixa salarial por atuação
salario_cargo_contagem = df.groupby(['Atuacao', 'Ano','Faixa salarial']).size()
salario_cargo_contagem=salario_cargo_contagem.reset_index(name='Contagem')
salario_cargo_contagem
```

```python
sns.violinplot(x="Ano", y="Faixa salarial", data=salario_cargo, palette="colorblind", bw=0.2, inner="box", cut=0, orient='h')
plt.show()
```
